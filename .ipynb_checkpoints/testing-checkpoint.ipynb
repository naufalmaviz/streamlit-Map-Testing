```python
import geopandas as gpd
import folium
from statistics import mean

All_Blocks = gpd.read_file(r'GeoJSON Files/All_Blocks.geojson')
All_Boreholes = gpd.read_file(r'GeoJSON Files/All_Boreholes.geojson')


# calculate the center of map
bounds = All_Blocks.total_bounds
x = mean([bounds[0], bounds[2]])
y = mean([bounds[1], bounds[3]])
location = (y, x)

map1 = folium.Map(location=location,
                    zoom_start=11, control_scale=300, tiles=None)

    # Put ESRI Satellite for the layer map
tile = folium.TileLayer(
                        tiles='https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
                        attr='Esri',
                        name='Esri Satellite',
                        overlay=False,
                        control=False
                        ).add_to(map1)

map1
```

```python
import folium
import rasterio
from branca.colormap import linear
```

```python
# Load the raster file
raster_file = "TIF Files\BATNAS_NORTH_ACEH_colours.tif"
src = rasterio.open(raster_file)

# Create a Folium map centered at the middle of the raster
m = folium.Map(location=[src.bounds.bottom + src.res[0]*src.height/2, src.bounds.left + src.res[1]*src.width/2], zoom_start=10)

# Add the raster to the map as a TileLayer
folium.raster_layers.ImageOverlay(
    image=src.read(1),
    bounds=[[src.bounds.bottom, src.bounds.left], [src.bounds.top, src.bounds.right]],
    opacity=0.7,
    interactive=True,
    cross_origin=False,
    colormap=lambda x: (1, 1, 1, x)
).add_to(m)

# Display the map
m
```
